```python
# split data into data_train and data_test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,shuffle=True,random_state=42)
```

### 4-Build Model

```python
# Build LogisticRegression model
from sklearn.linear_model import LogisticRegression
```

```python
# calling of LogisticRegression
Log_model=LogisticRegression(solver='liblinear',random_state=42)
```

```python
#fit / train linear LogisticRegression on train data (x_train, y_train)
Log_model.fit(x_train, y_train)
```

```python
#predict / test model on test data (x_test)
y_pred=Log_model.predict(x_test)
```

```python
print('first 10 value of y predict =',y_pred[:10])
print(' ')
print('first 10 value of y actual =',y_test.values[:10])
```

### 5-Model Evaluation

```python
from sklearn.metrics import accuracy_score
```

```python
# Calculate accuracy on training data
y_train_pred = Log_model.predict(x_train)
train_accuracy = accuracy_score(y_train, y_train_pred)
print(f'Training Accuracy: {train_accuracy}')
```

```python
# Calculate accuracy on test data
y_test_pred = Log_model.predict(x_test)
test_accuracy = accuracy_score(y_test, y_test_pred)
print(f'Test Accuracy: {test_accuracy}')
```

```python
# Print the Confusion Matrix 
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_test_pred)
print('Confusion matrix =', cm)
```

```python
# Plot the confusion matrix
plt.figure(figsize=(4,4))
sns.heatmap(cm,annot=True,fmt='.0f',cmap='Blues')
plt.title('confusion matrix')
plt.xlabel("predicted label")
plt.ylabel("actual label")
plt.show()
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_test_pred))
```

```python
# Applying 5-Fold Cross Validation

from sklearn.model_selection import cross_val_score

scores = cross_val_score(Log_model, x_train, y_train, cv = 5, scoring='accuracy')

print('Cross-validation scores:{}'.format(scores))
```

```python
# compute Average cross-validation score

print('Average cross-validation score: {}'.format(scores.mean()))
```

```python
# The roc_auc_score function measures the model's ability to distinguish between different classes
from sklearn.metrics import roc_auc_score

ROC_AUC = roc_auc_score(y_test, y_pred)

print('ROC AUC : {}'.format(ROC_AUC))
```
